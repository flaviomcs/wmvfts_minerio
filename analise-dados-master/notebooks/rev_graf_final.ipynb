```python
D_m = pd.to_datetime(D_m).dt.date
```

```python
def dt_ds(date):
  datas = date
  for dt in datas:
    ds = dt.strftime('%Y-%m-%d')
    datas_string.append(ds)
```

```python
datas_mes = []
```

```python
def transformar_datas_mensal(date):
  data_mes = date
  for datas in data_mes:
    data_split = datas.split('-')
    data_join = '-'.join(data_split[0:2])
    datas_mes.append(data_join)
```

```python
transformar_datas_mensal(datas_string)
```
