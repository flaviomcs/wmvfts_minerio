```python
def transformar_datas_mensal(date):
  data_mes = date
  for datas in data_mes:
    data_split = datas.split('-')
    if data_split[1]=="01":
      data_split[2]="31"
    if data_split[1]=="02":
      data_split[2]="28"
    if data_split[1]=="03":
      data_split[2]="31"
    if data_split[1]=="04":
      data_split[2]="30"
    if data_split[1]=="05":
      data_split[2]="31"
    if data_split[1]=="06":
      data_split[2]="30"
    if data_split[1]=="07":
      data_split[2]="31"
    if data_split[1]=="08":
      data_split[2]="31"
    if data_split[1]=="09":
      data_split[2]="30"
    if data_split[1]=="10":
      data_split[2]="31"
    if data_split[1]=="11":
      data_split[2]="30"
    if data_split[1]=="12":
      data_split[2]="31"
    data_join = '-'.join(data_split)
    datas_mes.append(data_join)
```
